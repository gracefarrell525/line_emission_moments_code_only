```python
'''KEYS'''

half_mass = 1 #0 for fullmass map, 1 for halfmass map
snapshot_key = 5 #1 for 151, 2 for 152, 5 for 155
tolerance = 1 #increase/decrease depending on wanted tolerance
emission = "CII_158mu" #line wanted for maps 
```

```python
''' READ IN OF HDF5 FILE '''

file_151 = '151_median_0001_moments_all_lines.hdf5'
file_152 = '152_median_0001_moments_all_lines.hdf5'
file_155 = '155_median_0001_moments_all_lines.hdf5'

if snapshot_key == 1:
    file_in = h5.File(file_151, "r")
    text = "Before"
if snapshot_key == 2:
    file_in = h5.File(file_152, "r")
    text = "During"
if snapshot_key == 5: 
    file_in = h5.File(file_155, "r")
    text = "After"

file_in.keys()
```
